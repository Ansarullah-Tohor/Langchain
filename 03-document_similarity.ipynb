```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')

```

```python
documents = [
   " Shakib Al Hasan is a Bangladeshi cricketer known for being one of the world’s best all-rounders.",
"Tamim Iqbal is a Bangladeshi opener famous for his aggressive stroke play and consistency.",
"Mushfiqur Rahim is known for his dependable batting and wicketkeeping under pressure.",
"Mahmudullah Riyad is a Bangladeshi middle-order batsman celebrated for his calm finishing and leadership.",
"Taskin Ahmed is a Bangladeshi fast bowler known for his pace, energy, and ability to take early wickets."
]

document_embedding= embedding.embed_documents(documents)

print(str(document_embedding))

query = 'tell me about taskin ahmed'
result=embedding.embed_query(query)
print(str(result))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

scores=cosine_similarity([result],document_embedding)[0]

print(scores)
```

```python
print(list(enumerate(scores)))
```

```python
index,score=sorted(list(enumerate(scores)),key=lambda x:x[1])[-1]
```

```python
print(query)
print(documents[index])
print("similarity score is:", score)
```
